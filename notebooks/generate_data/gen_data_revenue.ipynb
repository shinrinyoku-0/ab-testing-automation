```python
def sample_purchase_value(rng):
    bucket = rng.choice(
        ['small', 'medium', 'large'],
        p=[0.8, 0.15, 0.05]
    )

    if bucket == "small":
        return rng.uniform(10, 50)
    elif bucket == "medium":
        return rng.uniform(50, 200)
    else:
        return rng.uniform(200, 1000)
```
